Faça um programa que imprima os n primeiros números primos, onde n é informado pelo usuário.

```python
def primo(numero):
  for i in range(2,numero):
    if numero % i == 0:
      return False
  return True
n = int(input('Digite quantos números primos você quer: '))

i = 1
cont = 0
while cont < n:
  if primo(i) == True:
    cont +=1
    print(i,end=' ')
  i +=1
```
