```
%reset -f
hora=int(input("Insira quantas horas você tralbalhou este mês: "))
valorhora=float(input("Insira o valor da hora: "))
bruto=float(hora*valorhora)
print("Salário bruto: R$:", bruto)
sindicato=(3*bruto)/100
fgts=(11*bruto)/100
ir1=bruto
ir2=(5*bruto)/100
ir3=(10*bruto)/100
ir4=(20*bruto)/100
if bruto<=900:
  descontosir=0
  print("(-)IR(0%):R$ 0,00")
elif bruto<=1500:
  bruto2=((bruto-ir2)-bruto)
  descontosir=bruto2
  print("(-)IR(5%):R$ ",ir2)
elif bruto<=2500:
  bruto3=((bruto-ir3)-bruto)
  descontosir=bruto3
  print("(-)IRS(10%):R$ ",ir3)
elif bruto>2500:
  bruto4=((bruto-ir4)-bruto)
  descontosir=bruto4
  print("(-)IRS(20%):R$ ",ir4)
print("Sindicato: R$ ",sindicato)
print("FGTS:R$ ", fgts)
descontos=((descontosir-sindicato)*-1)
print("Total de descontos:R$ ",descontos)
liquido=((descontos-bruto)*-1)
print("Salário liquido:R$ ",liquido)
```

```
%reset -f 
a=float(input("Digite o valor de A: "))
b=float(input("Digite o valor de B: "))
c=float(input("Digite o valor de C: "))
delta=(b**2)-(4*a*c)
if a==0:
  print("A equação não é do segundo grau.")
elif delta<0:
  print("A equação não possui raízes, pois delta é negativo:", delta)
elif delta>0:
  x1=(-b)+(delta**(1/2))/(2*a)
  x2=(-b)-(delta**(1/2))/(2*a)
  print("A equação possui duas raízes:", x1, x2)
```
