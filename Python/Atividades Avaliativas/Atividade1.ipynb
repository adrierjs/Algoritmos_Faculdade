```python
%reset -f
lados=int (input("Insira quantos lados tem polígono: "))
cm=float (input("Insira a medida dos lado em cm: "))
if lados==3:
  area=(cm**2*3**(1/2))/4
  print("TRIÂNGULO", area,"cm²")
elif lados<3:
    print("NÃO É UM POLÍGONO")
elif lados==4:
  area=(cm**2)
  print("QUADRADO", area, "cm²")
elif lados==5:
  print("PENTÁGONO")
else:
  print("POLÍGONO NÃO IDENTIFICADO")
```
